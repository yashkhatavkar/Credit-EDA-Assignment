```python
def bar_comb_plot_chart(x,y):
    # 100% bar chart
    sns.despine
    plt.style.use('default')
    
    appdata = previous_application_data_final.pivot_table(values = 'SK_ID_CURR', 
                      columns = y, index = x, 
                      aggfunc='count')
    appdata=appdata.div(appdata.sum( axis = 1),axis = 'rows') * 100
    sns.set()
    appdata.plot(kind='bar',stacked=True,figsize=(16,6), style = 'default')
    plt.ylabel(f'{y}%')
    plt.xlabel(f'{x}')
    plt.title(f'Effect Of {x} on Loan Approval')
    plt.show()
```

```python
bar_comb_plot_chart('CODE_GENDER','NAME_CONTRACT_STATUS')
```

1. Gender has no relevenace in decision making process i.e (approval or rejection) of the application

3. The bank can focus more on the female applications in order to increase its topline and bottomline
